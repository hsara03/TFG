```python
import lightkurve as lk
import matplotlib.pyplot as plt
```

```python
target_star = "KIC 8462852"
```

```python
lc_collection= lk.search_lightcurve(target_star, mission="Kepler").download_all()
```

```python
if lc_collection is not None:
    lc = lc_collection.stitch()

    plt.figure(figsize=(10, 5))
    lc.plot()
    plt.title(f"Curva de luz de {target_star} (Kepler)")
    plt.xlabel("Tiempo (d√≠as)")
    plt.ylabel("Brillo normalizado")
    plt.show()

    # Guardamos los datos en un archivo CSV
    lc_df = lc.to_pandas()
    lc_df.to_csv("curva_luz_kepler.csv", index=False)
else:
    print("No se encontraron datos para la estrella seleccionada.")
```
